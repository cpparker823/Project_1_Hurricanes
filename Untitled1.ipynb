```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
import matplotlib.pyplot as plt
from sklearn import datasets
from pathlib import Path
import pprint
```

```python
Ocean_Temps = Path("Ocean_Temps.csv")
ocean_df = pd.read_csv(Ocean_Temps)
ocean_df
```

```python
# Convert 'time' column to datetime

ocean_df['time'] = pd.to_datetime(ocean_df['time'])


# Create a boolean mask to filter rows where the month is not September
mask = ocean_df['time'].dt.month != 9

# Filter the DataFrame using the boolean mask
sept_ocean_data = ocean_df[~mask]

sept_ocean_data_backup = sept_ocean_data

ocean_df['year'] = ocean_df['time'].dt.year

sept_ocean_data
```

```python
average_max_temp = sept_ocean_data.groupby(["year"]).mean().reset_index()

average_max_temp
```

```python
# Plot out ocean wind speed and ocean evap

x_values = sept_ocean_data["wind_speed_10m_max (mp/h)"]
y_values = sept_ocean_data["et0_fao_evapotranspiration (mm)"]
plt.scatter(x_values,y_values)
plt.xlabel('Ocean Wind Speed (Miles Per Hour)')
plt.ylabel('Ocean Evapotrasnpiration (Millimeters)')
plt.show()
```

```python
# add the linear regression equation and line 
x_values = sept_ocean_data["wind_speed_10m_max (mp/h)"]
y_values = sept_ocean_data["et0_fao_evapotranspiration (mm)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")

# Add the line equation
# plt.annotate(line_eq, (x_position, y_position), fontsize=12, color='red')

plt.annotate(line_eq,(20,.8),fontsize=18,color="red")
plt.xlabel('Ocean Wind Speed (Miles Per Hour)')
plt.ylabel('Ocean Evapotrasnpiration (Millimeters)')
print(f"The r-sqaured is: {rvalue**2}")
plt.show()

```
